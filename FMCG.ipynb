```python
for i in Columns_with_NA:
    values = final_dataset[i].astype(str).replace('nan','0')
    final_dataset.loc[:,[i]] = values

print(final_dataset['Stage2_QP3_High'].head())
print(final_dataset['Stage4_QP3_High'].head())
print(final_dataset.head())
```

```python
dummy_data = pd.get_dummies(final_dataset,drop_first=True)
dummy_data.head()
```

```python
dummy_data.shape
```

```python
X = dummy_data.iloc[:,:-1].values
y = dummy_data.iloc[:,-1].values
```

```python
#splitting the dataset into the training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=0)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
#feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
print(X_test)
```

```python
#Training the logistic Regression model on the training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
cf = classifier.fit(X_train,y_train)
print(cf.intercept_)
print(cf.coef_)
cf.predict_proba(X)
print(cf.score(X,y))
```

```python
#Predicting the test results
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
